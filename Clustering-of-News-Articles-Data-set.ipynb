```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
df=pd.read_json('NewsArticles.json')

vector=df['Vector']
vectordf=pd.DataFrame(list(vector))
```

```python
kmeans=KMeans(n_clusters=5,random_state=1)
pca=PCA(n_components=100,random_state=1)

kmeans.fit(vectordf)
kmeans.inertia_

sse1=kmeans.inertia_
pcavectordf=pca.fit_transform(vectordf)

kmeanspca=KMeans(n_clusters=5,random_state=1)
kmeanspca.fit(pcavectordf)

sse2=kmeanspca.inertia_
```

```python
df['WithoutPCA']=kmeans.labels_
df['WithPCA']=kmeanspca.labels_

withoutpca=df['WithoutPCA'].value_counts()
withpca=df['WithPCA'].value_counts()

mwopca=withoutpca.max()
iwopca=withoutpca.argmax()
mwpca=withpca.max()
iwpca=withpca.argmax()
```

```python
articleswithoutpca={}
articleswithpca={}

for x,y in df.groupby('WithoutPCA'):
    articleswithoutpca[x]=(' '.join(y['Preprocessed-Article']).lower().split()[:50])

for x,y in df.groupby('WithPCA'):
    articleswithpca[x]=(' '.join(y['Preprocessed-Article']).lower().split()[:50])
```

```python
output=[[np.round(sse1,2)],[np.round(sse2,2)],[iwopca],[mwopca],[iwpca],[mwpca],[articleswithoutpca[3][-1]],[articleswithpca[2][-1]]]
```

```python
import csv
with open('output.csv','w') as out:
    writer=csv.writer(out,delimiter=',')
    writer.writerows(output)
```
